```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns


df = pd.read_csv("player_data.csv", index_col = 0)
df.head()
df = df[["speed", "shoot", "pass", "dribble", "defense", "physical"]]
```

```python
plt.figure(figsize = (12, 9))
plt.scatter(df.iloc[:, 1], df.iloc[:, 3])
plt.xlabel('shoot')
plt.ylabel('dribble')
plt.title('shoot & dribble')
```

```python
scaler = StandardScaler()
std = scaler.fit_transform(df)
```

```python
pca = PCA()
pca.fit(std)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1, 7), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = "--")
plt.title('component')
plt.xlabel('Number of Component')
plt.ylabel('cumulative Explained Variance')
```

```python
pca = PCA(n_components = 4)
```

```python
pca.fit(std)
```

```python
pca.transform(std)
```

```python
scores_pca = pca.transform(std)
scores_pca
```

```python
wcss = []
for i in range(1, 21):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1, 21), wcss, marker = 'o', linestyle = "--")
plt.title('kmeans with PCA')
plt.xlabel('Number of Component')
plt.ylabel('cumulative Explained Variance')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 3)
```

```python
kmeans_pca.fit(scores_pca)
```

```python
df.head()
```

```python
df_pca_kmeans = pd.concat([df, pd.DataFrame(scores_pca)], axis = 1)
df_pca_kmeans.head()
df_pca_kmeans.columns.values[-4:] = ['c1','c2','c3','c4']
df_pca_kmeans['K-Means PCA'] = kmeans_pca.labels_
```

```python
df_pca_kmeans.head()
```

```python
df_pca_kmeans['Segment'] = df_pca_kmeans['K-Means PCA'].map({ 0 : 'first',
                                                            1 : 'second',
                                                            2 : 'third'})
```

```python
x_axis = df_pca_kmeans['c2']
y_axis = df_pca_kmeans['c1']
plt.figure(figsize = (10,8))
sns.scatterplot(x_axis, y_axis, hue = df_pca_kmeans['Segment'], palette = ['g', 'r', 'c'])
plt.title('clusters by PCA Components')
plt.show()
```
